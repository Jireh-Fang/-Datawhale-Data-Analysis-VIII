# 8.2 联合与合并数据集

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],
                   'data2':range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
df3 = pd.DataFrame({'lkey':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df4 = pd.DataFrame({'rkey':['a','b','d'],
                   'data2':range(3)})
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','b'],
                  'data1':range(6)})
df2 = pd.DataFrame({'key':['a','b','a','b','d'],
                  'data2':range(5)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,how='inner')
```

# 8.3 重塑和透视

## 8.3.1 使用多层索引进行重塑

```python
data = pd.DataFrame(np.arange(6).reshape((2,3)),
                   index=pd.Index(['Ohio','Colorado'],name='state'),
                   columns=pd.Index(['one','two','three'],
                                   name='number'))
data
```

```python
result = data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack(0)
```

```python
result.unstack('state')
```

```python
s1 = pd.Series([0,1,2,3],index=['a','b','c','d'])
s2 = pd.Series([4,5,6],index=['c','d','e'])
data2 = pd.concat([s1,s2],keys=['one','two'])
data2
```

```python
data2.unstack()
```

```python
data2.unstack().stack()
```

## 8.3.2 将‘长’透视为‘宽’

```python
data = pd.read_csv('examples/macrodata.csv')
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year,quarter=data.quarter,name='data')
colunmns = pd.Index(['realgdp','infl','unemp'],name='item')
data = data.reindex(columns=colunmns)
data.index = periods.to_timestamp('D','end')
ldata = data.stack().reset_index().rename(columns={0:'value'})
ldata[:10]
```

```python
pivoted = ldata.pivot('data','item','value')
pivoted
```

```python
pivoted = ldata.pivot('data','item')
pivoted[:5]
```

## 8.3.3 将“宽”透视为“长”

```python
df = pd.DataFrame({'key':['foo','bar','baz'],
                  'A':[1,2,3],
                  'B':[4,5,6],
                  'C':[7,8,9]})
df
```

```python
melted = pd.melt(df,['key'])
```

```python
melted
```

```python
reshaped = melted.pivot('key','variable','value')
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df,id_vars=['key'],value_vars=['A','B'])
```
